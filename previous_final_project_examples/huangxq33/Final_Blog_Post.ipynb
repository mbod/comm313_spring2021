## Looking at *A Love So Beautiful* from the Perspective of English-Speaking Audience: Text Analysis of English Audience Reviews

### Introduction

Near the end of 2017, a youth romance drama called *A Love So Beautiful* swept across the Chinese Internet. It tells the love story between a girl named Chen Xiao Xi and a boy named Jiang Chen. They have known each other since childhood and go to the same high school. As a handsome student with good grades, Jiang Chen is popular among the girls in school. Xiao Xi, on the other hand, does not study hard but loves drawing, and she is always so cheerful on expressing her admiration towards Jiang Chen. Jiang Chen seldom shows his affections toward Xiao Xi but often takes care of her. After graduating from high school, they go to the same college studying different majors. As they step into society and start working in different fields, some misunderstandings develop between the two. However, throughout nineteen years, their love and affections towards each other never change. At the end, they both learn to follow their hearts and fall in love with each other. Their high school classmates Lu Yang, Lin Jing Xiao, and Wu Bo Song accompany them throughout the years, too. While providing help and care to Jiang Chen and Xiao Xi, the three also have their love and friendship mingled with those of the couple.

When the official account of the drama posted the trailers and posters on Weibo (a Chinese social media platform similar to Twitter), #A Love So Beautiful# and related topics quickly went on the Weibo Hot Search. This Chinese drama was first released on Tencent Video in Mainland China on November 9th, 2017 and finished the last episode on December 7th, 2017. During its broadcasting period, discussions related to the drama spread across various Chinese social media platforms, such as Weibo and a review platform called Douban. There were also mash-up videos of the drama going viral on Bilibili, a video website where users can post their own video productions, including those made with clips from drama series.

I was in senior year in high school when *A Love So Beautiful* was being broadcasted. I remember that near the end of my senior year, I heard many classmates talking about this Chinese drama that it for once became a hot topic among certain groups of people in my school. When I posted about this phenomenon on my WeChat Moments, a friend of mine from another high school in Philadelphia commented that many of her friends were also crazy about this drama. As an immigrant student moving to the U.S. from China, I held a notion that Chinese dramas were not popular in the Western Hemisphere, so my U.S. friends' fondness of *A Love So Beautiful* really surprised me.

Therefore, I want to investigate how the English-speaking audience thinks about this popular Chinese drama. To do this, I decide to look at the English reviews for the drama posted in 2017 to 2019 and conduct text analysis on them. What do the viewers focus on when they watch the drama? What impresses them the most? Does the drama receive a good reputation outside of Mainland China?

### Approach to Analysis

We have learned several kinds of text analysis in class. For my research question, I think conducting **word and phrase frequency analysis**, **concordance analysis**, and **sentiment analysis** on the reviews would be useful.

In writings, words and phrases can reveal a lot about the writers, for example, their vocabulary levels, their personal thoughts, and their stances. Creating a word or phrase frequency list from reviews can provide an idea on what the viewers focus on or what impresses them the most when they watch the drama series. What aspect of the drama catches their attention? Does the drama remind them of their own experiences?

Taking a further step from word and phrase frequency analysis, concordance analysis allows me to look into details of the reviews and analyze the words of my interest. By investigating a word in context, I can discover and summarize patterns used with the word and discover meaning in these patterns. For example, interpreting the usage of a word in the reviews can help me understand what the reviewers are referring to with this specific word and why that might be important.

Lastly, conducting sentiment analysis is a good idea because I want to know the attitudes of the English-speaking audience towards the drama and whether or not the drama gains a good reputation in regions other than Mainland China. Sentiment analysis can quantify the types and degree of emotions that are expressed in the text. For example, if the result of the sentiment analysis shows that a review conveys positive/negative sentiments, I may be able to infer that the reviewer likes/dislikes or recommends/does not recommend the drama series.

### II. Concordance Analysis

Though word and phrase frequency analysis can give us a general idea of what the reviewers cover in their reviews, it cannot provide more detailed information about the reviewers' thoughts about the drama. For example, we see words such as *`school`* and *`cute`* that show up very often in the reviews, but we cannot tell from the frequency list that in what contexts these words appear and how these words are used. So for the next part, I am going to choose a word of interest from the word frequency list for each year and use concordance analysis to take a closer look at them. By investigating a specific word in context, we can summarize certain patterns used with the word and discover meaning in these patterns.

#### A. Investigating a word of interest for reviews in 2017

Let's recall the word frequency list (the most common 50 words) for reviews from 2017 and choose a word that we are interested in investigating.

![](images/word_freq_2017.png)

The word *`school`* appears 36 times. We already know that the drama is mainly about Jiang Chen and Xiao Xi's love story in school, but is there another "school" that the reviewers are referring to in their reviews? Would they write about their own love stories during their school life?

To examine the use of the word *`school`* by the reviewers, we need to:
1. create a list of tokens for the review texts from 2017
2. create a KWIC (key word in context) concordance object using the `make_kwic` function
3. obtain a random sample of 20 lines
4. examine the KWIC listing and make some observations using the `sort_kwic` and `print_kwic` functions

The `make_kwic` function makes a KWIC concordance object with a specified number of tokens to the left and right of the chosen word of interest. In my analysis, I need six words (i.e., six tokens) on the left and on the right of my chosen word of interest, so I let `win=6`.

In the `sort_kwic` function, I can specify the sorting order when I obtain the sample of lines. For example, with order `['R1']`, the sample will be presented according to the alphabetical order of the first word on the right side of the key word.

Here is a random sample of KWIC concordance objects for the word *`school`* in the reviews from 2017.

### III. Sentiment Analysis

For my research question, I am interested in looking at the overall polarity of a review (i.e., whether it is positive or negative). The overall polarity should give a sense about whether the drama receives a good reputation and it is the most straightforward indication of how the audience feels about the drama in terms of good or bad. The arousal, dominance, and emotion might tell us about the specific emotional feelings of the reviewer towards the drama (or even a particular scene in the drama), but those are not my focus. So, for sentiment analysis, I am going to look at the dimension of valence only.

The sentiment analysis tool I am going to use is VADER -- Valence Aware Dictionary and sEntiment Reasoner. The VADER lexicon has a list of items, each of which has a valence rating.

We need to set up the `SentimentIntensityAnalyzer` in order to use VADER.

I will point to a `SentimentIntensityAnalyzer` object with pointer `sid`. We can use this object along with the VADER lexicon to classify words (or tokens) in a text.

I will not go deep into examining the valence of every single word in each review text (since the reviews are long), but will apply the `polarity_scores` function of the `SentimentIntensityAnalyzer` to each review text. This function will take the proportion of words in the text that belongs to each valence category (*negative*, *neutral*, and *positive*) as their polarity score, and will calculate a compound score (between -1 and 1, where -1 means negative and 1 means positive) for the text.

### Conclusion

By conducting word and phrase frequency analysis, concordance analysis, and sentiment analysis on the MyDramaList reviews posted from 2017 to 2019 for the Chinese drama series *A Love so Beautiful*, we learn that the audience focus on various aspects when they watch a drama. For example, they pay attention to the plot, the setting, the characters, and even the music. We also see that different reviewers might use the same word in their reviews for different purposes: to refer to the elements in the drama or to something related to but not directly involved in the drama, to describe the story plot, and to describe the characters. The sentiment analysis reveals a potential phenomenon of drama-watching -- the audience may like the drama better when it is on air or recently finished, but such interest of the audience in the drama may decrease as the drama has ended and as time passes. Nevertheless, the general result of sentiment analysis suggests a similar attitude to what the website rating suggests, which is that the drama actually receives a good reputation among the English-speaking audience outside of Mainland China.

There are some limitations of this project. For example, MyDramaList.com has only 100 reviews posted in the time range from 2017 to 2019, which yields a small corpus; because of some technical issues, I did not use Rakuten Viki as my source for collecting reviews, which could have provided me a corpus of a larger size and might be useful when conducting the word and phrase frequency analysis and the concordance analysis since there would be a greater number of texts. Another limitation is that I did not look into other dimensions (e.g., emotion and arousal) of sentiment analysis, which might provide interesting insights into the reviewers' thoughts and opinions about the drama. It would also be a good idea to take a closer look at the specific words that yield a certain score in sentiment analysis. Since these are not within the range of my research question, I did not include them, but a project that examines these aspects would also be interesting.
