# Introduction
***
### While the novel coronavirus spread around the world, ravaging healthcare systems, economies, and daily living, most Americans turned to popular media outlets to garner live updates and information about the mysterious virus. With so much unknown about the virus, from how it spreads to whether people can develop immunity, these media outlets have become vital sources of coronavirus coverage, providing all of the information people need to form opinions about its severity, spread, and how they should act in response. The news and media outlets are always on and people are always watching.

### Since the 2016 election, the mainstream media has become highly politically polarized. Fox and CNN are two news outlets that have positioned themselves in opposite political spheres and are representative of this polarization. For example, Pew Data has shown that 75% of conservative republicans say that Fox is their main source of information. CNN and Fox are also two of the most watched news outlets in the U.S. for coronavirus coverage; both CNN and Fox have had between 2-5 million daily viewers. Given the increased attention placed on these news outlets during the coronavirus pandemic, as well as their starkly differing political affiliations, it is understandable that they would report coronavirus news differently, and that each of their viewers would therefore develop different opinions and viewpoints about the virus, its spread, and its severity. Recently, studies have been published that delve into this. For example, in April, a Pew Poll was published that highlighted differences in viewer responses to the coronavirus between different politically leaning news outlets. One thing the poll was found was that 80% of Fox News viewers believed that the media exaggerated the threat from coronavirus compared to 54% of CNN viewers. While this highlights how viewers of different politically leaning news outlets have formed different opinions and beliefs about the coronavirus, studies haven’t yet delved into the exact causes of these differing viewer reactions. Have these different politically leaning news outlets differed in their responses to the coronavirus? If so, how?

### I aimed to investigate this, by studying CNN and Fox news transcripts containing coronavirus coverage, and searching for linguistic differences between the two popular and different politically leaning news outlets. This would enable me to paint a picture of how CNN and Fox have differed in their coronavirus responses, and thus shed light on what may have caused differing viewer reactions to the coronavirus coverage of the different politically leaning news outlets.

# Data Collection
***

### The news transcript data for both CNN and Fox that I required was obtained from Lexis Nexis, an online library database that stores comprehensive news content, including news transcripts for all the major news outlets. I culled Fox and CNN news transcripts from January to the second week in April that contained the words ‘coronavirus’, ‘pandemic’, and/or ‘covid19.’ This provided me with almost 2,000 CNN and over 450 Fox news transcripts containing coronavirus coverage. Because news broadcasts tend to cover many different topics and contain tangents and side conversations between news anchors, I filtered each news transcript down to only the parts that directly addressed coronavirus and the above key words specifically. Doing this enabled me to create an accurate representation of each news transcript’s coronavirus coverage and response, by focusing directly on the areas of each news transcript that mentioned coronavirus.

## CNN and Fox Coronavirus Response Sentiment 
***
### Sentiment Analysis is a broad term used to describe natural language processing approaches used to quantify the types and degree of emotions expressed in text. Emotion, particularly the balance of negative and positive emotion, is a key element of a response to an event/topic. Two news outlets covering the same news topic may convey similar information with different emotion, and this ultimately can create major differences between the resulting emotions their viewers adopt themselves towards that topic. Additionally, different emotions may indicate that the two sources convey different content and information altogether while reporting on a given topic. Thus, conducting sentiment analysis on both Fox and CNN news broadcasts covering the coronavirus, analyzing sentiment trends over time, and comparing the results between the two news outlets, can reveal important differences between their overall responses to the coronavirus.

### I used Valence Aware Dictionary and sentiment Reasoner (VADER), a lexicon and rule-based sentiment analysis tool that is specifically attuned to sentiments expressed in social media. Specifically, I used a VADER algorithm to give each news transcript a compound sentiment score, ranging between -1 to 1, which represents the overall positive or negative direction and magnitude of the emotions of the text. I then created a sentiment score time series for both Fox and CNN as shown below, in order to observe differences in both the magnitudes and trajectories of the sentiment scores of their news transcripts over time.

### First, I found the average sentiment score of both Fox and CNN coronavirus responses between January and April. Despite coronavirus news broadcasts feeling very negative, both Fox and CNN appear to have sentiment scores that are only slightly negative on average, with Fox news transcripts being slightly more negative on average than CNN news transcripts (-0.18 vs. -0.09 average sentiment score). 

### As shown above, it appears that the sentiment of Fox news transcripts is more inconsistent compared with that of CNN news transcripts; Fox's daily news transcript sentiment varies much more than that of CNN's between January and April. In fact the average standard deviation of the daily sentiment scores of Fox and CNN are 0.275 and 0.199 respectively; there appears to be a meaningful difference between the standard deviations of the news outlets' sentiment. Inconsistency in sentiment could confuse viewers, and muddle the overall effectiveness and quality of a news outlet's response to a given topic. This data could indicate that Fox’s overall response to coronavirus has been all over the place/inconsistent relative to CNN’s. However, in order to support a broader conclusion like this, additional analyses delving into the consistency of each news outlet’s coronavirus response needed to be conducted.

# Fox and CNN Coronavirus Response Word Choice and Messaging Focus
***
### The above analyses delved into and revealed an important difference between Fox's and CNN's coronavirus response from different angles: consistency. Another important difference between the coronavirus responses of Fox and CNN could be their word choice and overall messaging focus/content. Keyness analysis is one tool that can help provide insight into this.

## Keyness Analysis
***
### Keyness Analysis is a statistical tool that is often used in order to identify significant differences between 2 corpora. Essentially, Keyness Analysis compares the normalized frequencies of linguistic items in two corpora. I used Keyness Analysis to compare normalized word frequencies between Fox and CNN news broadcasts, comparing word frequencies in CNN news broadcasts to those in Fox news broadcasts and vice versa. This helped to provide insights into the vocabulary, word choice, and messaging focus differences that exist between the two news outlets' coronavirus responses. The results from my Keyness Analysis are shown below.
<br>
<img src='images/figures/keyness_table.png' width='900'>

### From the table above, it appears that there are many interesting vocabulary differences between Fox and CNN news broadcasts. This helps to paint a picture of the messaging differences between the two news outlets' coronavirus responses. CNN appears to have focused its messaging around the United States, the spread of the coronavirus, and the results of this spread. Their response has contained significantly more coverage of the number of deaths in the U.S., the symptoms patients have suffered from, the number of confirmed cases, and the evolution of testing in the U.S. compared with Fox's response. Fox appears to have focused its messaging around China, focusing its coverage around both the country and Chinese people much more than CNN has. Fox mentions China significantly more than it mentions America (362 times versus 127 times), and mentions Chinese more than it mentions America as well (183 times versus 127 times). Fox mentions the words China and Chinese significantly more than CNN does. Moreover, it appears that Fox has made its messaging significantly more political than CNN has. Fox has mentioned the words democrats and Biden a total of 166 times which is not only significantly more than CNN has mentioned them, but is also more than Fox has mentioned America (127 times). Moreover, Fox appears to have focused a lot more attention around the response to coronavirus than CNN has. This is underscored by Fox's focus on the malaria drug hydroxychloroquine, which it has mentioned significantly more than CNN has. 

### Overall, it appears that there are certain words that distinguish CNN and Fox coronavirus responses from one another, and that vocabulary differences are strong enough to observe differences between the messaging focus of each news outlet's response. Exactly how important are these vocabulary differences between Fox and CNN coronavirus responses? Are these vocabulary differences strong enough to truly distinguish Fox and CNN coronavirus responses from one another and thus conclude that their messaging focus is different?

## Logistic Regression
***
### In order to answer the above questions I ran a logistic regression, a statistical tool used to create classification models to distinguish between two groups based on differentiating factors between the groups. I first found the 250 most common words that the news broadcasts use, and created a classification model on the entire news transcript data I had collected. I used the aforementioned words as explanatory variables and news outlet identity as a response variable, and this created a highly significant classification model as shown in the summary results below. This model is able to explain approximately 72.5% of the difference between Fox and CNN coronavirus responses (Pseudo R-squared = 0.725) and is highly significant (LLR p-value: 2.8176-79).
<br>
<img src='images/figures/first_model_summary.png' width='500'>


### I used this first model to do feature selection, narrowing down the list of differentiating words to those that had a very strong ability to differentiate between Fox and CNN coronavirus responses (words with p-values of less than 0.01 in the resulting model). This resulted in a list containing 36 words that highly distinguish between Fox and CNN coronavirus responses. A table containing the top 25 of these words can be found below.

### I then created a second model with the above words as explanatory variables and using cross-validation.
